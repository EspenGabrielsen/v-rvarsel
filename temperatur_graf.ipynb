```python
import pandas as pd
import sqlite3 
from matplotlib import pyplot as plt
```

```python
conn = sqlite3.connect("sample.db")
```

```python
df = pd.read_sql("select * from forcast",con=conn)
```

```python
df.time = pd.to_datetime(df.time)
```

```python
df_grp = df.groupby([pd.Grouper(key="time",freq='h'),"name"])["air_temperature"].mean().reset_index()
```

```python
plt.figure(figsize=(16,5))
x= df_grp.loc[df_grp.name == 'Korshamn',"time"]
y= df_grp.loc[df_grp.name == 'Korshamn',"air_temperature"]
plt.plot(x,y,label="Korshamn",lw=2)

x= df_grp.loc[df_grp.name == 'Lier',"time"]
y= df_grp.loc[df_grp.name == 'Lier',"air_temperature"]
plt.plot(x,y,label="Lier",lw=2)
plt.title("Time in UTC, temperature in celsius")
plt.legend(["Korshamn","Lier"])
plt.axhline(0,ls='--',color="black",alpha=.4)
```

```python
df_grp = df.groupby([pd.Grouper(key="time",freq='h'),"name"])["wind_speed"].mean().reset_index()
plt.figure(figsize=(16,5))
x= df_grp.loc[df_grp.name == 'Korshamn',"time"]
y= df_grp.loc[df_grp.name == 'Korshamn',"wind_speed"]
plt.plot(x,y,label="Korshamn",lw=2)

x= df_grp.loc[df_grp.name == 'Lier',"time"]
y= df_grp.loc[df_grp.name == 'Lier',"wind_speed"]
plt.plot(x,y,label="Lier",lw=2)
plt.title("Time in UTC, wind in knots")
plt.legend(["Korshamn","Lier"])
plt.axhline(0,ls='--',color="black",alpha=.4)
```

```python
df_grp = df.groupby([pd.Grouper(key="time",freq='h'),"name"])["precipitation"].mean().reset_index()
plt.figure(figsize=(16,5))
x= df_grp.loc[df_grp.name == 'Korshamn',"time"]
y= df_grp.loc[df_grp.name == 'Korshamn',"precipitation"]
plt.plot(x,y,label="Korshamn",lw=2)

x= df_grp.loc[df_grp.name == 'Lier',"time"]
y= df_grp.loc[df_grp.name == 'Lier',"precipitation"]
plt.plot(x,y,label="Lier",lw=2)
plt.title("Time in UTC, precipitation")
plt.legend(["Korshamn","Lier"])
plt.axhline(0,ls='--',color="black",alpha=.4)
```
